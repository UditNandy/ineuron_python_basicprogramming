### 1.Write a Python program to check if the given number is a Disarium Number?

```python
n=input("")
length=len(n)
n=int(n)
k=n
s=0
while(n!=0):
    rem=n%10
    s+=rem**length
    n//=10
    length-=1
if(s==k):
    print('Disarium')
else:
    print('Not Disarium')
```

### 2.Write a Python program to print all disarium numbers between 1 to 100?

```python
def isdisarium(n):
    length=len(str(n))
    k=n
    s=0
    while(n!=0):
        rem=n%10
        s+=rem**length
        n//=10
        length-=1
    if(s==k):
        return True
    else:
        return False
 
for i in range(1,101):
    if(isdisarium(i)):
        print(i,end=' ')
```
